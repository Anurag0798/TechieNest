```python
# There are two types of recommendation systems. They are
# Content-Based Recommender System
# Collaborative Recommender System

# Content-Based Recommender System
# A content-based recommender system works on the data generated from a user. The data can be generated either explicitly (like clicking likes) or implicitly (like clicking on links). This data will be used to create a user profile for the user which contain the metadata of the items user interacted. More the data it receives more accurate the system or engine becomes.

# Collaborative Recommender System
# A collaborative recommender system makes a recommendation based on how similar users liked the item. The system will group users with similar tastes. In addition to user similarity, recommender systems can also perform collaborative filtering using item similarity (like ‘Users who liked this item X also liked Y’).
```

```python
moviesmat = df4.pivot_table(index='user_id', columns='title', values='rating')
moviesmat
```

```python
starwars = moviesmat["Star Wars (1977)"]
starwars
```

```python
starwars_corr = moviesmat.corrwith(starwars)
starwars_corr
```

```python
ratings.head()
```

```python
ratings.sort_values('ratings_count', ascending=False).head()
```

```python
starwars_user_ratings = moviesmat['Star Wars (1977)']
starwars_user_ratings.head()
```

```python
similar_to_star_wars = moviesmat.corrwith(starwars_user_ratings)
similar_to_star_wars
```

```python
corr_starwars = pd.DataFrame(similar_to_star_wars, columns=['correlation'])
corr_starwars.dropna(inplace=True)
corr_starwars.head()
```

```python
corr_starwars.sort_values('correlation', ascending=False)
```

```python
corr_starwars = corr_starwars.join(ratings["ratings_count"])
corr_starwars.head()
```

```python
corr_starwars[corr_starwars["ratings_count"]>100].sort_values('correlation', ascending=False).head(6)
```

```python
liarliar_user_ratings = moviesmat['Liar Liar (1997)']

similar_to_liarliar = moviesmat.corrwith(liarliar_user_ratings)

corr_liarliar = pd.DataFrame(similar_to_liarliar, columns=['correlation'])
corr_liarliar.dropna(inplace=True)

corr_liarliar = corr_liarliar.join(ratings["ratings_count"])

corr_liarliar[corr_liarliar["ratings_count"]>100].sort_values('correlation', ascending=False).head(6)
```
