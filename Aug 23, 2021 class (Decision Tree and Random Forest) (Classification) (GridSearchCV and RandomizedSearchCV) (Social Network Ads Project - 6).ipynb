# GridSearchCV

```python
criterions = ['gini','entropy']
n_est = list(range(5,201,5))
min_splits = [2,4,6,8,10,14,18,20]

params = {'n_estimators':n_est, 'criterion':criterions,'min_samples_split':min_splits}
```

```python
params
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
g_model = GridSearchCV(RandomForestClassifier(), param_grid=params,verbose=3)
g_model.fit(xtrain,ytrain)
```

```python
g_model.best_estimator_
```

```python
g_model.best_params_
```

```python
g_model.best_score_
```

# RandomizedSearchCV

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
r_model = RandomizedSearchCV(RandomForestClassifier(), params, n_iter=10, scoring='accuracy', verbose=3, n_jobs=-1, random_state=5)
```

```python
r_model.fit(xtrain,ytrain)
```

```python
r_model.best_estimator_
```

```python
r_model.best_params_
```
